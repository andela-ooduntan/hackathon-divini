```python
for items in models:
    print(items)
    nb_model = models[items]
    nb_model.fit(X_train, y_train.ravel())
    prediction_test = nb_model.predict(X_test)
    prediction_train = nb_model.predict(X_train)
    print("Confusion matrix for test")
    print("{0}".format(metrics.confusion_matrix(y_test, prediction_test, labels=[0,1])))

    print("Classification Report")
    print("{0}".format(metrics.classification_report(y_test, prediction_test, labels=[0,1])))
    print(" ")
    print("Confusion matrix for train")
    print("{0}".format(metrics.confusion_matrix(y_train, prediction_train, labels=[0,1])))

    print("Classification Report")
    print("{0}".format(metrics.classification_report(y_train, prediction_train, labels=[0,1])))
```
